```python
import pandas as pd
import glob
from datetime import datetime 
```

```python
# Get a list of all the file paths that ends with .csv
all_files = glob.glob("crawl_13_features\*.csv")
extend_files = glob.glob("crawl_fed_and_gold\*.csv")
all_files.extend(extend_files)
all_files.remove('crawl_fed_and_gold\\Gold.csv')
all_files
```

```python
import pandas as pd

# Read the 'gold.csv' file
gold_data = pd.read_csv('crawl_fed_and_gold\\Gold.csv')  # Replace 'path/to/files/' with your directory

# Convert the 'date' column to datetime if it's not already in datetime format
gold_data['date'] = pd.to_datetime(gold_data['date'])

# Loop through each file and merge based on the 'date' column
for file_path in all_files:
    data = pd.read_csv(file_path)  # Read data from each file
    data['date'] = pd.to_datetime(data['date'])  # Convert 'date' column to datetime if needed

    # Merge data based on the 'date' column
    gold_data = pd.merge(gold_data, data, on='date', how='left')

gold_data.rename(columns = {'Open': 'GOLD_open',
                              'High': 'GOLD_high',
                              'Close': 'GOLD_close',
                              'Low': 'GOLD_low',
                              'date': 'Date'} ,inplace=True)
gold_data = gold_data.drop(columns = ['Unnamed: 0_x', 'Unnamed: 0_y'])

# Save the merged data to a new file or use it for further analysis
gold_data.to_csv('merged_data_complete.csv', index=False) 
```
